```python
import numpy as np
import matplotlib.pyplot as plt

name = ['APTES', 'Untreaded', 'GOPTS', 'Poly-lysine', 'UV-O3']
data=np.loadtxt("C:\PY\lab.txt")
ca_std=[]
ca_mean=[]

for i in (0,2,4,6,8):
    ca_mean.append(np.mean(180-data[i:i+2]))
    ca_std.append(np.std(data[i:i+2]))
    
sigma_lg = 71.99e-3 # N/m
    
plt.figure(figsize=((16,12)))
plt.bar(name,ca_mean,yerr=ca_std, color='orange', label='θ', width = 0.4)
plt.ylabel('Contat angle / °',fontsize=28)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0,fontsize=28)
plt.show()



W_A_mean = [] # J/m^2
W_A_std =[]

    
for i in (0,2,4,6,8):
    W_A_mean.append(sigma_lg*(np.cos(np.mean(180-data[i:i+2])/180*np.pi)+1))
    W_A_std.append(np.std(sigma_lg*(np.cos((180-data[i:i+2])/180*np.pi)+1)))
    
plt.figure(figsize=((16,12)))
plt.bar(name,W_A_mean,yerr=W_A_std,color='orange', label='work of adhesion',width=0.4)
plt.ylabel('work of adhesion  J / m^2',fontsize=28)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=2,fontsize=28)
plt.show()
```
