```python
print ('Problem #3:')

N_A = 6.02214082e+23 # Avogadro constant
bulk_density = 1.13e+6 # g/m^3
molar_mass = 45.05 # g/mol

# assume the shape of fomamide molecular is spherical.

molecular_volume = molar_mass/bulk_density/N_A # m^3/molecular
molecular_size = (molecular_volume*3/4/np.pi)**(1/3)*1e+10 # angstroms
molecular_density = bulk_density/molar_mass*N_A # moleculars/m^3

print ('(a): \n molecular_size ~ %.5f Angstroms (radius) \n molecular_density = %.5e molecular/m^3 \n' %(molecular_size,molecular_density))

B_vdw = 1.2e-76 # J*m^6 / molecules^2
A = (np.pi*molecular_density)**2*B_vdw # hamaker constant

print ('(b): \n Hamaker constant A = %.5e J \n' %A)

D_o = 2*molecular_size/1e+10 # Distance of closest approach: 2*radius

sigma = A/(24*np.pi*D_o**2)

print ('(c): \n Surface tension = %f N/m \n' %sigma)

print ('(d): \n 1. dispersion forces \n 2. permanent dipole \n 3. hydrogen bond')
```

```python
print ('Problem #4: \n')
print ('(a) and (b):')

sigma = 0.072 # N/m
radius = np.linspace(5e-9, 1e-5,1000) # radius/m
delta_P = 2*sigma/radius # Pa

plt.plot (radius,np.log10(delta_P/1000), label='ŒîP') 
plt.xlabel('radius / m')
plt.ylabel('ŒîP / kPa (in log10) ')

def collapse_point (r):
    return 2*sigma/r -101325

sol = optimize.root(collapse_point, [8e-6]) # find collapse point

plt.plot(np.linspace(5e-9,sol.x,100),np.linspace(1,1,100)*np.log10(101.325),color='green',label='stable')
plt.plot(np.linspace(sol.x,1e-5,100),np.linspace(1,1,100)*np.log10(101.325),'--',color='red',label='unstable')
plt.plot(sol.x,np.log10(101.325),'o',color='red', label='collapse point (1atm) \n r=%.5e m' %sol.x)

plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0)
plt.show()

sigma = 0.08255 # N/m for 6 mol/L NaCl

sol = optimize.root(collapse_point, [8e-7]) # find new collapse point
 
print ('(c): I would add some NaCl into the water, for 6 mol/L NaCl solution at 20¬∞C, the ùúé is 82.55 mN/m.',
       'Accordingly, the collapse point radius is %.5e m. Which is 14.7%% more than before.' %sol.x)
```
