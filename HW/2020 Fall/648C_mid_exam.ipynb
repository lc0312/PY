```python
import numpy as np
# clear night, 2.5m/s wind, it's F condition
u = 2.5 # m/s
Q = 40 # g/s
sigma_y = 7 # m
sigma_z = 4 # m
# assume no reflaction

def C (x):
    return Q/(2*np.pi*u*sigma_y*sigma_z)

print ('P1. %.5e μg/m^3' %(C(200)*1e+6))
```
