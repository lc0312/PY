```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data_excel = pd.DataFrame (pd.read_excel('D:\HW4_formulation data-1.xlsx')).to_numpy()


Name =['5.5-S','6.0-S','6.5-S','6.0-A','6.0-L','6.0-AS','6.0-AL','6.0']
data = []
day = [[0],[7],[14],[21],[28]]

for i in (2,9,16,23,30,37,44,51):
    data.append (data_excel[i:i+5,1:4].T)
```

```python
for name,i in zip(Name,range(8)):
    plt.plot (day,data[i][0],'o--', label=name)

plt.ylabel('%HMW')
plt.xlabel('Day')

plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0)
plt.show()
```

```python
for name,i in zip(Name,range(8)):
    plt.plot (day,data[i][1],'o--', label=name)

plt.ylabel('%Monomert')
plt.xlabel('Day')

plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0)
plt.show()
```

```python
for name,i in zip(Name,range(8)):
    plt.plot (day,data[i][2],'o--', label=name)

plt.ylabel('%Fragment')
plt.xlabel('Day')

plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0)
plt.show()
```

```python
regre = linear_model.LinearRegression()
hmw_pred = []
for name,i in zip(Name,range(8)):
    regre.fit(day,data[i][0])
    pred = regre.predict([[30.5]])
    hmw_pred.append(pred.item())
    
plt.bar(Name,hmw_pred, label='HMW')
plt.ylabel('%HMW / Month')
plt.legend(loc=0)
plt.show()

mon_pred = []
for name,i in zip(Name,range(8)):
    regre.fit(day,data[i][1])
    pred = regre.predict([[30.5]])
    mon_pred.append(pred.item())
    
plt.bar(Name,mon_pred, label='Monomer',color='green')
plt.ylabel('%Monomer / Month')
plt.ylim(bottom=85)
plt.legend(loc=0)
plt.show()

frag_pred = []
for name,i in zip(Name,range(8)):
    regre.fit(day,data[i][2])
    pred = regre.predict([[30.5]])
    frag_pred.append(pred.item())
    
plt.bar(Name,frag_pred, label='Fragment',color='greenyellow')
plt.ylabel('%Fragment / Month')
plt.ylim(bottom=2.0)
plt.legend(loc=0)
plt.show()
```
