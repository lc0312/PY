```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from numpy import linalg

T_f = 298  # k
rho_CP = 4000  # KJ/(m**3*K)
C_Af = 2.0  # kmol/m**3

delta_HR = -np.array ([50000,100000,220000,300000])
theta = np.array ([3500,700,90,40])
pzt = np.array([221,222,223,224]) # positions of plot

for delta_HR,theta,pzt in zip(delta_HR,theta,pzt): # loop with different parameters

    def find_T_from_theta(T): # give a theta and find Temperature.
        
        k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
        
        # combine two steady_state equation in to one: bring C_A=C_Af/(k*theta+1) into second one
        return (T_f-T)/theta + (-delta_HR)*k*C_Af/(rho_CP*(k*theta+1)) # combined euqation
       
    # find the root of equation ( find the limit Temperature or  maximum Temperature)
    sol = optimize.root(find_T_from_theta, [500])
    
    # define the range of temperature for plot, sol.x is the limit Temperature, 0.25 is dT
    T = np.arange (T_f, sol.x, 0.25)
    
    k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
    
    def theta(T):
        # this is the equation of theta, derived from combined euqation
        return ((T_f - T) * rho_CP) / (k * (delta_HR * C_Af - rho_CP * (T_f - T))) 

    plt.subplot(pzt) 
    plt.tight_layout() 
    plt.plot(theta(T), T, label=r'$\Delta H_R=%s kJ/mol $' % (delta_HR / 1000)) # plot theta vs Temperature
    plt.xlabel(r'$\Theta /min $')
    plt.ylabel('Temperature/K')
    plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
    plt.minorticks_on()
    plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
    plt.legend(loc=7,fontsize= 'x-small')
#plt.savefig ('Steady-state Theta-Temperature ', dpi=400)
plt.show ()
```

```python
delta_HR = -np.array([50000, 100000, 220000, 300000])
theta = np.array([3500, 700, 90, 50])
pzt = np.array([221, 222, 223, 224])

for delta_HR, theta, pzt in zip(delta_HR, theta, pzt):
    def find_T_from_theta(T):
        k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
        return (T_f - T) / theta + (-delta_HR) * k * C_Af / (rho_CP * (k * theta + 1))

    sol = optimize.root(find_T_from_theta, [500])
    T = np.arange(T_f, sol.x, 0.25)
    k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
    
    def theta(T):
        return ((T_f - T) * rho_CP) / (k * (delta_HR * C_Af - rho_CP * (T_f - T)))
    
    # equation of X (conversion), X=C_A/C_Af, C_A=C_Af/(k*theta+1)
    X=k*theta(T)/(k*theta(T)+1)

    plt.subplot(pzt)
    plt.tight_layout()
    plt.plot(theta(T), X, label=r'$\Delta H_R=%s kJ/mol $' % (delta_HR / 1000)) # plot theta vs X (conversion)
    plt.xlabel(r'$\Theta /min $')
    plt.ylabel('COnversion')
    plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
    plt.minorticks_on()
    plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
    plt.legend(loc=7, fontsize='x-small')
#plt.savefig ('Steady-state Theta-Conversion', dpi=400)
plt.show()
```

```python
T_f = 298  # k
rho_CP = 4000  # KJ/(m**3*K)
C_Af = 2.0  # kmol/m**3
delta_HR=-220000
Theta_r = [0,250] # define the range of theta for plot
Theta = np.linspace (Theta_r[0],Theta_r[1],5*Theta_r[1]) # split one theta into 5 part to find temperature
UA_V = 250 # pick 250 as value of UA/V

for T_a in(300,330,360):
    T_record = [] # use a vector to store the temperature of each theta
    for theta in (Theta):
        def find_T_from_theta(T):

            k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
            return (T_f - T) + ((-delta_HR)*k*C_Af*theta) / (rho_CP * (k * theta + 1)) + theta*UA_V*(T_a-T)/rho_CP

        sol = optimize.root(find_T_from_theta, [325]) # for each theta find the temperature
        T_record.append(sol.x) # record temperature of each theta
    plt.plot (Theta,T_record, label=r'T_a=%.0f K' %(T_a)) # plot theta vs temperature
plt.xlabel('Theta/min')
plt.ylabel('Temperature/K')
plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=1)
plt.show()
```

```python
Theta_r = [0,500] # define the range of theta for plot
Theta = np.linspace (Theta_r[0],Theta_r[1],5*Theta_r[1]) # split  one theta into 5 part to find temperature
UA_V = 250 # pick 250 as value of UA/V

for T_a in (360,330,300): # define different Ta value to plot
    X_record = [] # use a vector to store the conversion of each theta
    for theta in (Theta):     
        def find_T_from_theta(T):

            k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
            return (T_f - T) + ((-delta_HR)*k*C_Af*theta) / (rho_CP * (k * theta + 1)) + theta*UA_V*(T_a-T)/rho_CP

        sol = optimize.root(find_T_from_theta, [300]) # for each theta find the temperature
        T = sol.x # temperature
        k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
        X_record.append (k*theta/(k*theta+1)) # record each X of different theta and temperature
    
    plt.plot (Theta,X_record, label=r'T_a=%.0f K' %(T_a)) # plot theta vs X
plt.xlabel('Theta/min')
plt.ylabel('Conversion')
plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc=0)
plt.show()
```

```python
t = [0, 350] # define time for ODE
delta_HR = -220000
UA_V = 250
T_a = 330

for theta in (30, 35, 72): # set different theta for plot
    def df_dt(t, f):
        C_A, T = f # define two variable of ODE equation,one is concentration C_A, another is temperature T
        k = 4.56 * 10 ** 8 * np.exp(-8000 / T)
        
        #two equations for ODE, dCA/dt and dT/dt
        return [((C_Af - C_A) / theta) - k * C_A,
                (T_f - T) / theta + (-delta_HR) * k * C_A/rho_CP + UA_V*(T_a - T)/ rho_CP]

    # ODE command, RK45 method, stepsize is 0.2 minute.
    sol = solve_ivp(df_dt, [t[0], t[1]], [C_Af, T_f], t_eval=np.linspace(t[0], t[1], 5 * t[1]))
    
    # plot the time vs temperature in different theta
    plt.plot(sol.t, sol.y[1], label=r'$\Theta =%s min $' % (theta))

plt.xlabel('Time/min')
plt.ylabel('Temperature/K')
plt.grid(b=1, which='major', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=1, which='minor', linestyle='-', alpha=0.2)
plt.legend(loc='best')
plt.show()
```
