```python
# Q1
'''
From figure we can mesure that slope of curve A is 0.0467, R_ct=46.7 ohm.
Then use equation RT/(nFi_o)=R_ct to calculate  exchange current i_o.
''' 
# Assume T=298K, n=1.
R = 8.314 #J/(mol*K)
T = 298 #K
n = 1
F = 96485 #C/mol
R_ct = 46.7 #ohm
i_o = R*T/(n*F*R_ct)

print ('exchange current i_o =%f mA' %(i_o*1000))
```
